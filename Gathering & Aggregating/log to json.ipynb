```python
listings_df=pd.DataFrame()
for i in starting_points:
    file_name=i+".json"
    json_data=open(file_name).read()
    data = json.loads(json_data)
    my_dict ={}
    for key,value in data['data']['Search']['find'].items():
        my_dict[key]=value
    if(i==0):
        listings_df=pd.DataFrame(my_dict['listings'])
    else:
        listings_df =pd.concat([listings_df, pd.DataFrame(my_dict['listings'])], ignore_index=True)
    
```
